### 1. Bubble Sort
**Pseudo-code:**
```
procedure bubbleSort(list)
   n = length(list)
   for i from 0 to n-1
      for j from 0 to n-i-2
         if list[j] > list[j+1]
            swap(list[j], list[j+1])
end procedure
```

**Analisis Kompleksitas:**
*   **Worst Case:** O(n²) - terjadi saat data terurut terbalik.
*   **Best Case:** O(n) - jika menggunakan flag untuk mengecek apakah ada pertukaran.
*   **Average Case:** O(n²)

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

data = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort Result:", bubble_sort(data.copy()))
````

### 2. Selection Sort
**Pseudo-code:**
```
procedure selectionSort(list)
   n = length(list)
   for i from 0 to n-1
      min_idx = i
      for j from i+1 to n-1
         if list[j] < list[min_idx]
            min_idx = j
      swap(list[i], list[min_idx])
end procedure
```

**Analisis Kompleksitas:**
*   **Worst, Best, & Average Case:** O(n²) - karena selalu ada dua loop bersarang terlepas dari kondisi data.

````python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print("Selection Sort Result:", selection_sort(data.copy()))
````

### 3. Insertion Sort
**Pseudo-code:**
```
procedure insertionSort(list)
   for i from 1 to length(list)-1
      key = list[i]
      j = i - 1
      while j >= 0 and list[j] > key
         list[j + 1] = list[j]
         j = j - 1
      list[j + 1] = key
end procedure
```

**Analisis Kompleksitas:**
*   **Worst Case:** O(n²) - data terurut terbalik.
*   **Best Case:** O(n) - data sudah terurut.
*   **Average Case:** O(n²)

````python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

print("Insertion Sort Result:", insertion_sort(data.copy()))
````

### 4. Merge Sort
**Pseudo-code:**
```
procedure mergeSort(list)
   if length(list) > 1
      mid = length(list) / 2
      left = list[0..mid]
      right = list[mid..n]
      mergeSort(left)
      mergeSort(right)
      merge(list, left, right)
end procedure
```

**Analisis Kompleksitas:**
*   **Worst, Best, & Average Case:** O(n log n) - algoritma ini menggunakan pendekatan Divide and Conquer yang sangat efisien untuk dataset besar.

````python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

print("Merge Sort Result:", merge_sort(data.copy()))
````
